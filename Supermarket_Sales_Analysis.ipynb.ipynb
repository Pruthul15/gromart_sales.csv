```python
# Step 1: Import the pandas library
import pandas as pd

```

```python
# Step 2: Load the dataset into a DataFrame
df = pd.read_csv('supermarket_sales.csv')

# View the first 5 rows
df.head()
```

```python
# Step 3: Create a new column 'total' to calculate total sale per transaction
df['total'] = df['unit price'] * df['Quantity']
df[['Branch', 'unit price', 'Quantity', 'total']].head()
```

```python
df.columns

```

```python
df[['branch', 'unit_price', 'quantity', 'total']].head()
```

```python
# Group by Branch and sum the Total column
branch_revenue = df.groupby('Branch')['Total'].sum().sort_values(ascending=False)

# Display the result
branch_revenue
```

```python
import matplotlib.pyplot as plt

# Plot the revenue by branch
branch_revenue.plot(kind='bar', title='Total Revenue by Branch', figsize=(8,5))
plt.xlabel('Branch')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.tight_layout()
plt.savefig('images/branch_revenue.png')  # Save chart to images folder
plt.show()
```

```python
# Save the result to a CSV file
branch_revenue.to_csv('outputs/branch_revenue.csv')
```

```python
# Group the sales data by 'Product line'
# Then, calculate the total revenue for each product category
product_revenue = df.groupby('Product line')['Total'].sum().sort_values(ascending=False)

# Show the result
product_revenue
```

```python
# Import matplotlib (if not already imported)
import matplotlib.pyplot as plt

# Create a bar chart to show revenue by product category
product_revenue.plot(kind='bar', title='Total Revenue by Product Category', figsize=(10,6))
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)  # Rotate labels to fit nicely
plt.grid(True)
plt.tight_layout()
plt.savefig('images/product_revenue.png')  # Save chart to 'images' folder
plt.show()
```

```python
# Save the product revenue summary as a CSV file
product_revenue.to_csv('outputs/product_revenue.csv')
```

```python
# Group the data by 'Payment' method
# Then calculate total revenue for each payment type
payment_revenue = df.groupby('Payment')['Total'].sum().sort_values(ascending=False)

# Show the result
payment_revenue
```

```python
# Plot a bar chart to compare revenue by payment method
payment_revenue.plot(kind='bar', title='Total Revenue by Payment Type', figsize=(8,5))
plt.xlabel('Payment Method')
plt.ylabel('Total Revenue')
plt.xticks(rotation=0)  # Keep labels straight
plt.grid(True)
plt.tight_layout()
plt.savefig('images/payment_revenue.png')  # Save chart to 'images' folder
plt.show()
```

```python
# Save the result to outputs folder
payment_revenue.to_csv('outputs/payment_revenue.csv')
```

```python
# Convert the 'Date' column to actual datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Create a new column with the day name (e.g., Monday, Tuesday)
df['Day'] = df['Date'].dt.day_name()
df[['Date', 'Day']].head()
```

```python
# Group by day and sum the total revenue
day_revenue = df.groupby('Day')['Total'].sum()

# Sort the days in real week order
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_revenue = day_revenue.reindex(day_order)

# Show result
day_revenue
```

```python
# Plot revenue by day
day_revenue.plot(kind='bar', title='Total Revenue by Day of the Week', figsize=(10,5))
plt.xlabel('Day of Week')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.tight_layout()
plt.savefig('images/revenue_by_day.png')
plt.show()
```

```python
# Convert 'Time' column to datetime format (just time, not full datetime)
df['Time'] = pd.to_datetime(df['Time'])
```

```python
# Convert 'Time' column to datetime using specific format (Hour:Minute)
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M')
```

```python
# Extract the hour
df['Hour'] = df['Time'].dt.hour
df[['Time', 'Hour']].head()
```

```python
# Compare average spending between 'Member' and 'Normal' customers
avg_spending_by_type = df.groupby('Customer type')['Total'].mean().sort_values(ascending=False)
avg_spending_by_type
```

```python
# Plot average spending by customer type
avg_spending_by_type.plot(kind='bar', title='Avg Spending: Member vs Normal', figsize=(6,4))
plt.xlabel('Customer Type')
plt.ylabel('Average Spend')
plt.grid(True)
plt.tight_layout()
plt.savefig('images/avg_spending_by_customer_type.png')
plt.show()
```

```python
# Compare average spending between Male and Female customers
avg_spending_by_gender = df.groupby('Gender')['Total'].mean().sort_values(ascending=False)
avg_spending_by_gender
```

```python
# Plot average spending by gender
avg_spending_by_gender.plot(kind='bar', title='Avg Spending: Male vs Female', figsize=(6,4))
plt.xlabel('Gender')
plt.ylabel('Average Spend')
plt.grid(True)
plt.tight_layout()
plt.savefig('images/avg_spending_by_gender.png')
plt.show()
```

```python
# Save both results to outputs folder
avg_spending_by_type.to_csv('outputs/avg_spending_by_customer_type.csv')
avg_spending_by_gender.to_csv('outputs/avg_spending_by_gender.csv')
```

```python
# Import seaborn for better visuals
import seaborn as sns

# Create a scatter plot: Rating vs Total Spend
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Rating', y='Total')
plt.title('Customer Rating vs Total Spending')
plt.xlabel('Customer Rating')
plt.ylabel('Total Sale Amount')
plt.grid(True)
plt.tight_layout()
plt.savefig('images/rating_vs_total.png')
plt.show()
```

```python
# Advanced scatter plot: Color points based on sale amount
plt.figure(figsize=(10,6))

# Use seaborn scatterplot with color gradient
sns.scatterplot(
    data=df,
    x='Rating',
    y='Total',
    hue='Total',         # Color based on how big the sale is
    palette='Reds',      # Use red shades
    size='Total',        # Bigger dot = bigger sale
    sizes=(10, 200),     # Range of dot sizes
    alpha=0.6,           # Make points a little transparent
    edgecolor='black'    # Add border to make dots pop
)

plt.title('Customer Rating vs Total Spending (Colored by Sale Amount)', fontsize=14)
plt.xlabel('Customer Rating')
plt.ylabel('Total Sale Amount')
plt.legend(title='Total Sale', loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.savefig('images/rating_vs_total_colored.png')
plt.show()
```
